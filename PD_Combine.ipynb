```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
a = pd.read_csv('Injury_Number_by_Type_Sex_Age_2014_2015.csv')
b = pd.read_csv('Injury_Number_by_Type_Sex_Age_2016.csv')
c = pd.read_csv('Injury_Number_by_Type_Sex_Age_2017.csv')
d = pd.read_csv('Injury_Number_by_Type_Sex_Age_2018.csv')
e = pd.read_csv('Injury_Number_by_Type_Sex_Age_2012_2013.csv')
f = pd.read_csv('Injury_Number_by_Type_Sex_Age_2011.csv')
g = pd.read_csv('Injury_Number_by_Type_Sex_Age_2009_2010.csv')
h = pd.read_csv('Injury_Number_by_Type_Sex_Age_1999_2008.csv')
```

```python
z = h.append(c).append(b).append(a).append(e).append(f).append(g).append(d)
```

```python
z.info()
```

```python
z.describe()
```

```python
z = z.reset_index()
z = z.drop(['index'],axis =1)
z.head()
```

```python
for i in range(1999,2019):
    for j in ['0-1','1-4','0-4']:
        for t in ['Male','Female','Persons']:
            total = z.Value[(z.Sex == t) & (z.Age == j) & (z.Year==i)].sum()
            if total != 0:
                for each in z.Type.unique():
                    print('The percentage of', each,"of", t,'for',j,'in',i,'is',(z.Value[(z.Sex == t) & (z.Age == j) & (z.Year==i)& (z.Type == each)]/total))
```

```python
z.Value[z.Type == 'Fall'] = 0.25 * (z.Value[z.Type == 'Fall'])
```

```python
y = z[z.Sex == 'Persons']
y = y.drop(['Sex'],axis = 1)
y.to_csv('Injury_Number_by_Type_Age_1999_2018.csv',index=None)
```
