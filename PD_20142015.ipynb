### Drown

```python
drown = pd.read_excel('20142015.xlsx',sheet_name='Drown')
```

```python
drown.columns
```

```python
for i in ['0-1','1-4']:
    for j in [2014,2015]:
        l1 = [[i,'Persons', j,drown.Value[(drown['Age'] == i) & (drown['Year'] == j)].sum()]]
        l1 = pd.DataFrame(l1,columns=['Age','Sex','Year','Value'])
        print(l1)
        drown = drown.append(l1)
```

```python
for i in ['Male','Female']:
    for j in [2014,2015]:
        l2 = [['0-4',i, j,drown.Value[(drown['Sex'] == i) & (drown['Year'] == j)].sum()]]
        l2 = pd.DataFrame(l2,columns=['Age','Sex','Year','Value'])
        print(l2)
        drown = drown.append(l2)
```

```python
for i in [2014,2015]:
    l3 = [['0-4','Persons',i,drown.Value[(drown['Sex']=='Persons') & (drown['Year']==i)].sum()]]
    l3 = pd.DataFrame(l3,columns=['Age','Sex','Year','Value'])
    print(l3)
    drown = drown.append(l3)
```

```python
drown
```

```python
drown['Type'] = 'Drown'
```
