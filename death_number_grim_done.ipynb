```python
b = a.reset_index()
```

```python
del b['index']
```

```python
b.iloc[5,0]
```

```python
for i in range(b.shape[0]):
    b.iloc[i,0] = re.sub(" \(.*?\)","",b.iloc[i,0])
```

```python
b
```

```python
b = b.fillna(0)
```

```python
b.head()
```

```python
b['population'] = 0
```

```python
b.columns = ['cause_of_death','year','sex','death_number','death_rate','population']
```

```python
b.death_number = b.death_number.astype(float)
b.death_rate = b.death_rate.astype(float)
```

```python
b['population'] = round((100000*b['death_number']/b['death_rate']),1)
```

```python
death_rates = b[b['cause_of_death']=='All causes combined']
```
