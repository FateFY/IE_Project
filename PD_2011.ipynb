### Drown

```python
drown = pd.read_excel('2011.xlsx',sheet_name='Drown')
drown.columns
```

```python
for i in ['0-1','1-4']:
    l2 = [[i,'Persons',drown.Value[drown.Age==i].sum(),2011]]
    l2 = pd.DataFrame(l2,columns=['Age','Sex','Value','Year'])
    drown = drown.append(l2)
    
l3 = [['0-4','Persons',drown.Value[drown.Sex=='Persons'].sum(),2011]]
l3 = pd.DataFrame(l3,columns =['Age','Sex','Value','Year'])
drown = drown.append(l3)

for i in ['Male','Female']:
    l4 =[['0-4',i,drown.Value[drown.Sex==i].sum(),2011]]
    l4 = pd.DataFrame(l4,columns=['Age','Sex','Value','Year'])
    drown = drown.append(l4)
```

```python
drown['Type'] = 'Drown'
```
